# Linear Regression

Linear Regression: LR used to understand the linear dependency between dependent(target) and independent(features) variables.
Our Research question was "Can we use movie budget to predict movie revenue?"
target = movie revenue
feature = movie budget
Data was collected from "https://www.the-numbers.com/"

```python
# Import Libraries
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Read the Data
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
X = DataFrame(data, columns = ['production_budget_usd'])
y = DataFrame(data, columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X,y, alpha = 0.3)
plt.title('Production vs Revenue')
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope Coefficient

```python
regression.coef_  # beta 1
```

beta 1 has positive value, this means budget and revenue has positive relationship. one unit increase in budget (X) can increase the 3.111 times of revenue (y). $1 increase in budget can bring $3.111 revenue.

```python
regression.intercept_  # intercept beta 0
```

This value litrely tells us a movie with $0 budget would actually lose $7236192.73.. So thats a big problematic. that's why we are just focused on the prediction at extreme end. 

```python
plt.figure(figsize = (10,6))
plt.scatter(X,y, alpha = 0.3)
plt.plot(X, regression.predict(X), color = 'red', linewidth = 4)
plt.title('Production vs Revenue')
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

R-Square

```python
regression.score(X,y) # R-Squared value
```

this value explains the amount variations in revenue (dependent variable) by the budget (independent variable). almost 55% variability explain by the budget in revenue from this model.
